```python
def sigmoid(x):
    return 1 / (1 + np.exp(-1 * x))

def sigmoid_der(x):
    return x * (1.0 - x)
```

```python
def train(x, y, w1, w2, bias):
    temp = []
    second_layer = np.dot(x, w1) + bias[:3]
    y_second = [sigmoid(x_second) for x_second in second_layer]
    
    output = np.dot(y_second, w2) + bias[3]
    y_second.append(sigmoid(output))
    error = abs((y - y_second[-1]) / y)
    
    q13 = sigmoid_der(y_second[-1]) * (y -y_second[-1])
    for k in range(3):
        b = q13 * y_second[k]
        temp.append(w2[k] + b)
    bias[3] += b
    
    for item in range(3):
        q = sigmoid_der(y_second[item]) * q13 * w2[item]
        for k in range(42):
            b = q * x[k]
            w1[k, item] += b
        bias[item] += b
    
    w2 = []
    w2.extend(temp)
    
    return w1, w2, bias, y_second[-1], error, output
```

```python
def online_train(x, y, w1, w2, bias, epoches):
    y_train = []
    delta_train = []
    epoch = 1
    outputs = []
    
    while epoch <= epoches:
        temp_1, temp_2 = [], []
        for k in range(len(x)):
            w1, w2, bias, result, error, output = train(x[k], y[k], w1, w2, bias)
            outputs.append(output)
            temp_1.append(result)
            temp_2.append(error)
        if epoch % 10 == 0 or epoch == 1:
            print(y[:5])
            print(outputs[:5])
            outputs = []
            print("Epoch: ", epoch)
            print("Error: ", statistics.mean(temp_2))
        y_train.append(temp_1)
        delta_train.append(temp_2)
        epoch += 1
        
        
    
    np.savetxt("w1.csv", w1, delimiter=",")
    np.savetxt("w2.csv", w2, delimiter=",")
    np.savetxt("bias.csv", bias, delimiter=",")
    
    return w1, w2, bias
```

```python
def recognize(x, w1, w2, bias):
    print('Recognize: ')
    count_1 = 0
    count_2 = 0
    for i in range(len(x)):
        second_layer = np.dot(x[i], w1) + bias[:3]
        y_second = [sigmoid(x_second) for x_second in second_layer]
        output = np.dot(y_second, w2) + bias[3]
        res = sigmoid(output)
        y_second.append(res)
        
        if res > 0.5:
            count_1 += 1
        else:
            count_2 += 1
    
    print('Count normal: ', count_1)
    print('Count land attacks: ', count_2)
    
    return y_second[-1]
```

```python
# Generate weights
np.random.seed(1)
w_1 = np.random.uniform(-3, 3, (42, 3))    
w_2 = np.random.uniform(-3, 3, (3, ))
bias = np.random.uniform(-3, 3, (4, ))
```

```python
res = online_train(normalized_df, output_data, w_1, w_2,  bias, 100)
```
